### 4. Identificação e Tratamento de Outliers
Identificar outliers visualmente com boxplots.

Tratar outliers, se necessário, usando técnicas como a regra do IQR ou transformações.

```python
# O IQR de salary_in_usd será de Q3 - Q1 = 70905.75
df_data_scientist[['salary_in_usd']].plot(kind='box');

q1_salary = np.percentile(df_data_scientist['salary_in_usd'], 25)
q3_salary = np.percentile(df_data_scientist['salary_in_usd'], 75)
iqr_salary = q3_salary - q1_salary
iqr_salary
```

```python
q1_salary -(iqr_salary * 1.5) 
```

```python
q3_salary + (iqr_salary * 1.5) 
```

Neste caso, vamos identifacar o os outliers onde qualquer dado que estiver fora de Q3+1.5*IQR ou Q1-1.5*IQR será considerado um outlier.

Minimo = 12735.625

Máximo = 83641.375

```python
# Identificando os outliers através de funções

df_sem_outliers = df_data_scientist[(df_data_scientist['salary_in_usd'] <= 296358) & (df_data_scientist['salary_in_usd'] >= 12735)]
df_sem_outliers
```

### Distribuição salarial
Nesta etapa podemos ver a distribuição dos salários para sabermos a média e mediana salarial dos profissionais da área considerando o salário convertido para dólar

A média salarial pode ser facilmente distorcida caso algum salário esteja muito acima ou abaixo em relação aos demais salários. No entanto, a mediana mostra de uma forma mais explícida qual a maior frequência em que o valor é pago.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['salary_in_usd'], bins=30, kde=True)

media_salarial = df['salary_in_usd'].mean()
mediana_salarial = df['salary_in_usd'].median()
print(f'A média salarial de profissionais da área é de U${round(media_salarial, 2)} por ano')
print(f'A mediana salarial de profissionais da área é de U${round(mediana_salarial, 2)} por ano')

plt.axvline(x=media_salarial, color='red', linestyle='dotted', label='Média Salarial')
plt.axvline(x=mediana_salarial, color='green', linestyle='solid', label='Mediana Salarial')
plt.title('Distribuição Salarial Geral')
plt.xlabel('Salário')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

### Tendência temporal
Nesta etapa podemos verificar a tendência salarial de acordo com o tempo.

Com isto, podemos validar a valorização destes profissionais ao longo do tempo e entender se a tendência está ascendente, descentente ou não há uma tendência definida em relação ao período passado.
