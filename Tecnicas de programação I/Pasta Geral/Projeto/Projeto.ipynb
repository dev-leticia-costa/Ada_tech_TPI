Contagem (count): Número de observações não nulas.

Valor Mínimo (min): O valor mínimo na coluna.
25º Percentil (25%): Primeiro quartil, indica o valor abaixo do qual 25% dos dados estão.
50º Percentil (50%): Mediana ou segundo quartil, o valor abaixo do qual 50% dos dados estão.
75º Percentil (75%): Terceiro quartil, o valor abaixo do qual 75% dos dados estão.
Valor Máximo (max): O valor máximo na coluna.

Desvio padrão: medida relativa, considerado de acordo com a média. Se o desvio padrão é significativamente menor que a média, pode ser considerado baixo. Se for significativamente maior, pode ser considerado alto. Essa avaliação também pode variar de acordo com o contexto.

### Distribuição salarial
Nesta etapa podemos ver a distribuição dos salários para sabermos a média e mediana salarial dos profissionais da área considerando o salário convertido para dólar

A média salarial pode ser facilmente distorcida caso algum salário esteja muito acima ou abaixo em relação aos demais salários. No entanto, a mediana mostra de uma forma mais explícida qual a maior frequência em que o valor é pago.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['salary_in_usd'], bins=30, kde=True)

media_salarial = df['salary_in_usd'].mean()
mediana_salarial = df['salary_in_usd'].median()
print(f'A média salarial de profissionais da área é de U${round(media_salarial, 2)} por ano')
print(f'A mediana salarial de profissionais da área é de U${round(mediana_salarial, 2)} por ano')

plt.axvline(x=media_salarial, color='red', linestyle='dotted', label='Média Salarial')
plt.axvline(x=mediana_salarial, color='green', linestyle='solid', label='Mediana Salarial')
plt.title('Distribuição Salarial Geral')
plt.xlabel('Salário')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```
